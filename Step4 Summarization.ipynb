# Summarize

```python
s_data = pd.read_csv(b_ac_st_t_0219 ,header = 0 ,sep = ',' ,encoding="utf-8-sig", index_col=0)
```

```python
s_data .info()
```

```python
s_data['Cyc_Balance_Decile_rank']=pd.qcut(s_data['Cyc_Balance_Grade'],10,labels=False) 
s_data['Cyc_Balance_Std'] = s_data['Cyc_Balance_Std'].fillna(0)
s_data.loc[s_data.Cyc_Balance_Std == 0.0, 'Balance_MayBe'] = 'New/Small' 
s_data.loc[(s_data.Cyc_Balance_Std <= s_data.Cyc_Balance_Std.mean())& (s_data.Cyc_Balance_Std != 0), 'Balance_MayBe'] = 'Low Variance' 
s_data.loc[(s_data.Cyc_Balance_Std >= s_data.Cyc_Balance_Std.mean())& (s_data.Cyc_Balance_Std != 0), 'Balance_MayBe'] = 'Height Variance'
s_data.loc[s_data.Cyc_Balance_Std == 0.0, 'Balance_Variance_Score'] = 0.0 
s_data.loc[(s_data.Cyc_Balance_Std <= s_data.Cyc_Balance_Std.mean())& (s_data.Cyc_Balance_Std != 0), 'Balance_Variance_Score'] = 2.0 
s_data.loc[(s_data.Cyc_Balance_Std >= s_data.Cyc_Balance_Std.mean())& (s_data.Cyc_Balance_Std != 0), 'Balance_Variance_Score'] = 1.0 
s_data['Payment_Decile_rank']=pd.qcut(s_data['Payment_Grade'],10,labels=False)
s_data['Payment_Std'] = s_data['Payment_Std'].fillna(0)
s_data.loc[s_data.Payment_Std == 0.0, 'Payment_MayBe'] = 'New/Small' 
s_data.loc[(s_data.Payment_Std <= s_data.Payment_Std.mean()) & (s_data.Payment_Std != 0), 'Payment_MayBe'] = 'Low Variance' 
s_data.loc[(s_data.Payment_Std >= s_data.Payment_Std.mean()) & (s_data.Payment_Std != 0), 'Payment_MayBe'] = 'Height Variance' 
s_data.loc[s_data.Payment_Std == 0.0, 'Payment_Variance_Score'] = 0.0
s_data.loc[(s_data.Payment_Std <= s_data.Payment_Std.mean()) & (s_data.Payment_Std != 0), 'Payment_Variance_Score'] = 2.0 
s_data.loc[(s_data.Payment_Std >= s_data.Payment_Std.mean()) & (s_data.Payment_Std != 0), 'Payment_Variance_Score'] = 1.0
s_data['Cash_Decile_rank']=pd.qcut(s_data['Cash_Grade'],10,labels=False)
s_data['Cash_Std'] = s_data['Cash_Std'].fillna(0)
s_data.loc[s_data.Cash_Std == 0.0, 'Cash_MayBe'] = 'New/Small' 
s_data.loc[(s_data.Cash_Std <= s_data.Cash_Std.mean()) & (s_data.Cash_Std != 0), 'Cash_MayBe'] = 'Low Variance' 
s_data.loc[(s_data.Cash_Std >= s_data.Cash_Std.mean()) & (s_data.Cash_Std != 0), 'Cash_MayBe'] = 'Height Variance' 
s_data.loc[s_data.Cash_Std == 0.0, 'Cash_Variance_Score'] = 0.0 
s_data.loc[(s_data.Cash_Std <= s_data.Cash_Std.mean()) & (s_data.Cash_Std != 0), 'Cash_Variance_Score'] = 2.0 
s_data.loc[(s_data.Cash_Std >= s_data.Cash_Std.mean()) & (s_data.Cash_Std != 0), 'Cash_Variance_Score'] = 1.0 
s_data['Sum_of_Decile_rank'] = s_data['Cyc_Balance_Decile_rank'] + s_data['Payment_Decile_rank'] + s_data['Cash_Decile_rank']
s_data['Sum_of_Variance_Score'] = s_data['Balance_Variance_Score'] + s_data['Payment_Variance_Score'] + s_data['Cash_Variance_Score'] 
s_data['Total'] = s_data['Sum_of_Decile_rank'] + s_data['Sum_of_Variance_Score']
```

```python
s_data.sort_values(['Total'],ascending=False)
```
