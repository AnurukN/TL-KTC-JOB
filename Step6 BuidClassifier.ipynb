## Data preparation

```python
#Load all data.
df_all = pd.read_csv(all_data, encoding = 'utf-8-sig', index_col = 0)
df_all = df_all.dropna(axis = 0)
df_for_predict = df_all.copy()
```

```python
df = pd.read_csv(t_data, encoding = 'utf-8-sig', index_col = 0)
df = df.drop(columns=['Customer_Status'])
```

```python
#Cat code
YN_types = ['Y', 'N'] #yes/no
FM_types = ['Female','Male'] #Gender
M_types = ['Single','Married','Unknown']
Oc_types = df_all.Occupation.unique() # Occupation
```

```python
Oc_types
```

```python
df_for_predict['Gender'] = df_for_predict.Gender.astype("category", categories=FM_types).cat.codes
df_for_predict['Marital_status'] = df_for_predict.Marital_status.astype("category", categories=M_types).cat.codes
df_for_predict['Mobilephone_fg'] = df_for_predict.Mobilephone_fg.astype("category", categories=YN_types).cat.codes
df_for_predict['Officephone_fg'] = df_for_predict.Officephone_fg.astype("category", categories=YN_types).cat.codes
df_for_predict['Homephone_fg'] = df_for_predict.Homephone_fg.astype("category", categories=YN_types).cat.codes
df_for_predict['Occupation'] = df_for_predict.Occupation.astype("category", categories=Oc_types).cat.codes
df_for_predict['Is_Staff'] = df_for_predict.Is_Staff.astype("category", categories=YN_types).cat.codes
df_for_predict['Is_KTC_Credit_Card_Holder'] = df_for_predict.Is_KTC_Credit_Card_Holder.astype("category", categories=YN_types).cat.codes
```

```python
df['Gender'] = df.Gender.astype("category", categories=FM_types).cat.codes
df['Marital_status'] = df.Marital_status.astype("category", categories=M_types).cat.codes
df['Mobilephone_fg'] = df.Mobilephone_fg.astype("category", categories=YN_types).cat.codes
df['Officephone_fg'] = df.Officephone_fg.astype("category", categories=YN_types).cat.codes
df['Homephone_fg'] = df.Homephone_fg.astype("category", categories=YN_types).cat.codes
df['Occupation'] = df.Occupation.astype("category", categories=Oc_types).cat.codes
df['Is_Staff'] = df.Is_Staff.astype("category", categories=YN_types).cat.codes
df['Is_KTC_Credit_Card_Holder'] = df.Is_KTC_Credit_Card_Holder.astype("category", categories=YN_types).cat.codes
df['Curr_Cpgn_Response_Flag'] = df.Curr_Cpgn_Response_Flag.astype("category", categories=YN_types).cat.codes
```

```python
from sklearn.model_selection import train_test_split
X = df[Feature_col] #df_train_under[Feature_col]
Y = df[Target_col] #df_train_under[Target_col]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

New_feature = SelectKBest(f_classif, k=47).fit(X_train, Y_train)
```

```python
mask = New_feature.get_support() #list of booleans
new_features = [] # The list of your K best features
feature_names = list(X_train.columns.values)

for bool, feature in zip(mask, feature_names):
    if bool:
        new_features.append(feature)
```

```python
new_features
```

```python
#reduce columns of a train-test set
X_train_new = pd.DataFrame(X_train, columns=new_features)
X_test_new = pd.DataFrame(X_test, columns=new_features)
```

# Buid Model

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import RFE, VarianceThreshold
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline

pipe = make_pipeline(
    RFE(estimator=ExtraTreesClassifier(criterion="gini", max_features='auto', n_estimators=100), step=0.9000000000000001),
    VarianceThreshold(threshold=0.2),
    VarianceThreshold(threshold=0.2),
    KNeighborsClassifier(n_neighbors=25, p=2, weights="distance")
)
estimator = pipe
estimator.fit(X_train_new,Y_train)

Y_predict_train = estimator.predict(X_train_new)
Y_predict_test = estimator.predict(X_test_new)
```

```python
estimator
```

```python
from sklearn.metrics import accuracy_score, f1_score
print("Accuracy and F1 with repect to the train set:.. ", accuracy_score(Y_train, Y_predict_train) * 100, f1_score(Y_train, Y_predict_train))

print("Accuracy and F1 with repect to the test set:.. ", accuracy_score(Y_test, Y_predict_test) * 100, f1_score(Y_test, Y_predict_test))
```

```python
from sklearn.metrics import confusion_matrix
print("The Confusion Metrix of feature of train set")
unique_label = np.unique(Y_train)
print(pd.DataFrame(confusion_matrix(Y_train,
                                    Y_predict_train,labels=unique_label), 
                index=['true:{:}'.format(x) for x in unique_label], 
                columns=['pred:{:}'.format(x) for x in unique_label]))
```

```python
from sklearn.metrics import confusion_matrix
print("The Confusion Metrix of feature of test set")
unique_label = np.unique(Y_test)
print(pd.DataFrame(confusion_matrix(Y_test,
                                    Y_predict_test,labels=unique_label), 
                index=['true:{:}'.format(x) for x in unique_label], 
                columns=['pred:{:}'.format(x) for x in unique_label]))
```
